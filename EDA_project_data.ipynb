```python
import numpy as np
import pandas as pd
import re
import seaborn as sns 
import matplotlib.pyplot as plt
#from PIL import Image
from IPython.display import Image
%matplotlib inline 
sns.set(color_codes=True)
```

```python
data = pd.read_csv('driver_imgs_list.csv')
data = data.iloc[:,:-1]
```

```python
data.head()
```

```python
data.describe()
```

```python
real_names = {'c0': 'safe driving',
'c1': 'texting - right',
'c2': 'talking on the phone - right',
'c3': 'texting - left',
'c4': 'talking on the phone - left',
'c5': 'operating the radio',
'c6': 'drinking',
'c7': 'reaching behind',
'c8': 'hair and makeup',
'c9': 'talking to passenger'}
```

```python
data['classname'] = data['classname'].map(real_names) 
```

```python
print(data['classname'].unique())
```
